# Resolution of the exercise

In my pursuit of constructing the most effective machine learning model, I recognized the importance of selecting an optimal test size which is a critical factor in model evaluation. My approach was methodical: I established a procedure to assess the performance of various classifiers with different proportions of test data.

To achieve this, I defined a function named evaluate_test_sizes. This function iterates through a list of potential test sizes, splits the dataset accordingly, and then applies cross-validation using GridSearchCV to fine-tune the hyperparameters for each model. Specifically, I focused on the F1 score a balanced metric for model performance and computed the mean F1 score across the cross-validation folds, along with its standard deviation. This standard deviation serves as a measure of the model's stability: the lower it is, the more reliable the model's performance.

```python
def evaluate_test_sizes(model, param_grid, X, y, test_sizes):
    mean_scores = {}
    std_scores = {}

    for size in test_sizes:
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=size)
        grid = GridSearchCV(model, param_grid, cv=10, scoring='f1', verbose=1)
        grid.fit(X_train, y_train)
        cv_scores = grid.cv_results_['mean_test_score']
        mean_scores[size] = np.mean(cv_scores)
        std_scores[size] = np.std(cv_scores)

    best_size = max(mean_scores, key=mean_scores.get)
    return best_size, mean_scores, std_scores

test_sizes = [0.2, 0.3, 0.4]
```

I applied this evaluation process to several models: Logistic Regression, Decision Tree, K Nearest Neighbors (KNN), Random Forest, and Multilayer Perceptron (MLP). For each model, I defined a grid of hyperparameters to explore. The output of my function provided me with the best test size for each model, based on the highest mean F1 score.

Through this systematic analysis, I was able to determine the test size that yielded the best performance for each type of classifier. This informed my subsequent modeling decisions, ensuring that each classifier was evaluated in the most favorable context.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import f1_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier

# Load the data
X = df.drop('Target', axis=1)
y = df['Target'].apply(lambda x: 1 if x == 'R' else 0)

# Define basic parameter grids for each model for this evaluation. Next, in the next step the hyperparameter setting will be developed.
param_grid_logreg = {'C': [1, 10], 'penalty': ['l2']}
param_grid_decision_tree = {'max_depth': [10, 20], 'min_samples_split': [2, 5]}
param_grid_knn = {'n_neighbors': [3, 5], 'weights': ['uniform', 'distance']}
param_grid_rf = {'n_estimators': [50, 100], 'max_depth': [10, 20]}
param_grid_mlp = {'hidden_layer_sizes': [(50,), (100,)], 'activation': ['relu', 'tanh']}

# Evaluate test sizes for each model
best_size_logreg, mean_scores_logreg, std_scores_logreg = evaluate_test_sizes(LogisticRegression(), param_grid_logreg, X, y, test_sizes)
best_size_decision_tree, mean_scores_decision_tree, std_scores_decision_tree = evaluate_test_sizes(DecisionTreeClassifier(), param_grid_decision_tree, X, y, test_sizes)
best_size_knn, mean_scores_knn, std_scores_knn = evaluate_test_sizes(KNeighborsClassifier(), param_grid_knn, X, y, test_sizes)
best_size_rf, mean_scores_rf, std_scores_rf = evaluate_test_sizes(RandomForestClassifier(), param_grid_rf, X, y, test_sizes)
best_size_mlp, mean_scores_mlp, std_scores_mlp = evaluate_test_sizes(MLPClassifier(), param_grid_mlp, X, y, test_sizes)

print("Best test size for Logistic Regression:", best_size_logreg, "with mean:", mean_scores_logreg, "and std deviation:", std_scores_logreg)
print("Best test size for Decision Tree:", best_size_decision_tree, "with mean:", mean_scores_decision_tree, "and std deviation:", std_scores_decision_tree)
print("Best test size for KNN:", best_size_knn, "with mean:", mean_scores_knn, "and std deviation:", std_scores_knn)
print("Best test size for Random Forest:", best_size_rf, "with mean:", mean_scores_rf, "and std deviation:", std_scores_rf)
print("Best test size for MLP:", best_size_mlp, "with mean:", mean_scores_mlp, "and std deviation:", std_scores_mlp)
```

After determining the optimal test size for each model, my next step was to fine-tune the hyperparameters and validate the models' performance. To this end, I crafted the function tune_and_evaluate, which not only iteratively adjusts the models' hyperparameters using GridSearchCV but also cross-validates the performance to ensure robustness.

Within this function, for each iteration up to 20, I split the dataset using the previously determined best test size, ensuring that the model is trained and tested on appropriately sized datasets. The GridSearchCV then meticulously searches through the hyperparameter space, finding the best combination based on the F1 score, which I use as my model's scoring function due to its balance between precision and recall, crucial for our imbalanced dataset.

Once the best parameters are identified for each iteration, I evaluate and store the model's performance metrics: the F1 score and accuracy on both training and test sets. This iterative process results in a list of the best parameters and their corresponding validation scores, which I then analyze to find the most commonly selected hyperparameters — my model's sweet spot.

In the final stages of this function, I train the model one last time with these most common hyperparameters on the last training set. Then, I evaluate its performance on the last test set to calculate the final F1 score and accuracy, providing a realistic assessment of the model's predictive capabilities. The standard errors for the validation scores, F1 scores, and accuracies are also computed to gauge the stability and reliability of the model's performance across iterations.

The output of this function is a fine-tuned, validated model along with a comprehensive set of performance metrics and the final train/test data splits, ready for any further diagnostic analysis, such as constructing a confusion matrix. By employing this systematic approach, I ensure that the chosen model is not only statistically sound but also practical for real-world predictions.

```python
from sklearn.metrics import accuracy_score, f1_score
from collections import Counter
from sklearn.model_selection import train_test_split, GridSearchCV

def tune_and_evaluate(model, detailed_param_grid, X, y, best_test_size, iterations=20):
    best_params_list = []
    validation_scores = []
    train_f1_scores = []
    test_f1_scores = []
    train_accuracies = []
    test_accuracies = []
    final_model = None
    final_X_train, final_X_test, final_y_train, final_y_test = None, None, None, None  # To store the sets of the last iteration to make a confusion matrix with the best model.

    for _ in range(iterations):
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=best_test_size)
        grid = GridSearchCV(model, detailed_param_grid, cv=10, scoring='f1', verbose=1)
        grid.fit(X_train, y_train)

        final_X_train, final_X_test, final_y_train, final_y_test = X_train, X_test, y_train, y_test

        # Recording the best parameters and validation score
        best_params_list.append(grid.best_params_)
        validation_scores.append(grid.best_score_)

        # Training and test scores
        best_model = grid.best_estimator_
        y_pred_train = best_model.predict(X_train)
        y_pred_test = best_model.predict(X_test)
        train_f1_scores.append(f1_score(y_train, y_pred_train))
        test_f1_scores.append(f1_score(y_test, y_pred_test))
        train_accuracies.append(accuracy_score(y_train, y_pred_train))
        test_accuracies.append(accuracy_score(y_test, y_pred_test))

    # Determine the most common parameters
    most_common_params = Counter([tuple(params.items()) for params in best_params_list]).most_common(1)[0][0]

    # Evaluate the model with the most common parameters
    final_model = model.set_params(**dict(most_common_params))
    final_model.fit(final_X_train, final_y_train)
    final_f1_score = f1_score(final_y_test, final_model.predict(final_X_test))
    final_accuracy = accuracy_score(final_y_test, final_model.predict(final_X_test))

    # Calculating the standard error
    validation_error = (np.var(validation_scores) / len(validation_scores)) ** 0.5
    train_f1_error = (np.var(train_f1_scores) / len(train_f1_scores)) ** 0.5
    test_f1_error = (np.var(test_f1_scores) / len(test_f1_scores)) ** 0.5
    train_accuracy_error = (np.var(train_accuracies) / len(train_accuracies)) ** 0.5
    test_accuracy_error = (np.var(test_accuracies) / len(test_accuracies)) ** 0.5

    return final_model, {
        "validation_score": np.mean(validation_scores),
        "validation_error": validation_error,
        "train_f1_score": np.mean(train_f1_scores),
        "train_f1_error": train_f1_error,
        "test_f1_score": final_f1_score,
        "test_f1_error": test_f1_error,
        "train_accuracy": np.mean(train_accuracies),
        "train_accuracy_error": train_accuracy_error,
        "test_accuracy": final_accuracy,
        "test_accuracy_error": test_accuracy_error
    }, dict(most_common_params), best_test_size, final_X_train, final_X_test, final_y_train, final_y_test
```

To visually compare the F1 scores of my machine learning models, I utilized a custom function named plot_model_performance. This function generates a bar chart with error bars, allowing me to easily identify which model performs best on the test set considering both the F1 score and its variability. I used a color palette from Seaborn to enhance the visual appeal. By calling this function with my models' metrics, I could efficiently display their performance and make informed decisions about model selection.

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_model_performance(metrics_dict, metric_key, error_key, title, palette):
    """
    Function to plot the performance of models based on a specific metric and its error.

    :param metrics_dict: Dictionary with metrics for each model.
                         Key is the model name, value is another dictionary with metrics.
    :param metric_key: Key for the specific metric to plot.
    :param error_key: Key for the error of the specific metric.
    :param title: Title of the plot.
    :param palette: Color palette to use for the plot.
    """
    model_names = list(metrics_dict.keys())
    mean_scores = [metrics[metric_key] for metrics in metrics_dict.values()]
    score_errors = [metrics[error_key] for metrics in metrics_dict.values()]
    colors = sns.color_palette(palette, len(model_names))

    plt.figure(figsize=(10, 6))
    plt.bar(model_names, mean_scores, yerr=score_errors, color=colors, capsize=5)
    plt.xlabel('Models')
    plt.ylabel(metric_key)
    plt.title(title)
    plt.show()

# Creating a dictionary with the metrics for each model
metrics_dict = {
    'Logistic Regression': metrics_logreg,
    'Decision Tree': metrics_decision_tree,
    'KNN': metrics_knn,
    'Random Forest': metrics_random_forest,
    'MLP': metrics_mlp
}

# Calling the function to plot the F1 score with a nice color palette
plot_model_performance(metrics_dict, 'test_f1_score', 'test_f1_error', 'Comparison of F1 Score Among Models', 'Set2')
```

In my analysis, consistency across training, testing, and validation scores is paramount. To identify the most stable model, I developed the find_best_model function. This function computes the standard deviation of the F1 scores from training, testing, and validation to gauge each model's consistency. I then select the model with the least variation, indicating reliability. The best model is determined not only by high accuracy but also by how consistently it performs across different data sets.

```python
def find_best_model(metrics_dict):
    """
    Encuentra el mejor modelo basado en la menor variación entre las puntuaciones de entrenamiento,
    prueba y validación.

    :param metrics_dict: Diccionario con las métricas de cada modelo.
    :return: Nombre del mejor modelo.
    """
    min_variation = float('inf')
    best_model = None

    for model_name, metrics in metrics_dict.items():
        train_score = metrics['train_f1_score']
        test_score = metrics['test_f1_score']
        val_score = metrics['validation_score']
        variation = np.std([train_score, test_score, val_score])

        if variation < min_variation:
            min_variation = variation
            best_model = model_name

    return best_model

# Identify the best model
best_model_name = find_best_model(metrics_dict)
print("Best Model:", best_model_name)
```

To evaluate the predictive accuracy of my selected model, I created a confusion matrix, which provides insight into the true versus predicted classifications. By plotting the matrix, I can visually assess the number of correct predictions against the misclassifications.
```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Function to generate the confusion matrix
def plot_confusion_matrix(model, X_test, y_test):
    y_pred = model.predict(X_test)
    cm = confusion_matrix(y_test, y_pred)

    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.show()

# Dictionary that maps the model names to their corresponding variables
model_info = {
    'Logistic Regression': (final_model_logreg, best_params_logreg, best_test_size_logreg, X_test_logreg, y_test_logreg),
    'Decision Tree': (final_model_decision_tree, best_params_decision_tree, best_test_size_decision_tree, X_test_decision_tree, y_test_decision_tree),
    'KNN': (final_model_knn, best_params_knn, best_test_size_knn, X_test_knn, y_test_knn),
    'Random Forest': (final_model_random_forest, best_params_random_forest, best_test_size_random_forest, X_test_random_forest, y_test_random_forest),
    'MLP': (final_model_mlp, best_params_mlp, best_test_size_mlp, X_test_mlp, y_test_mlp)
}
```
